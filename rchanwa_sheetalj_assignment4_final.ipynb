```python
##importing the packages.
import numpy as np
import matplotlib.pyplot as plt
import gym
from gym import spaces
```

```python
class GridEnvironment(gym.Env):
  metadata = { 'render.modes': []}

  def __init__(self):
    self.observation_space = spaces.Discrete(16)
    self.action_space = spaces.Discrete(4)
    self.max_timesteps = 20
    self.timestep = 0
    self.start_position = [0, 0]   #start and goal positions 
    self.agent_position = [0, 0]
    self.goal_position = [3, 3]
    ##positive rewards
    self.reward1_position = [0, 2]
    self.reward2_position = [2, 2]
    ##negative rewards 
    self.negative_reward1 = [0, 3]
    self.negative_reward2 = [2, 3]

    self.state = np.zeros((4, 4))  ## 4 states are defined.
    self.state[tuple(self.agent_position)] = 1
    self.state[tuple(self.goal_position)] = 0.7


  def reset(self):
    self.timestep = 0
    self.agent_position = [0, 0]
    self.states = np.zeros((4, 4))
    self.reward1_done = False
    self.reward2_done = False
    return self.agent_position

  def step(self, action):
    self.temp = self.agent_position.copy()
    if action == 0:
      self.agent_position[0] += 1 # agent can go down,up,left,right so we have 4 actions defined
    elif action == 1:
      self.agent_position[0] -= 1 
    elif action == 2:
      self.agent_position[1] += 1 
    elif action == 3:
      self.agent_position[1] -= 1 
    
    self.agent_position = np.clip(self.agent_position, 0, 3)
    reward = 0
    
    if (self.agent_position == self.negative_reward1).all():
      reward = -10
      self.agent_position = np.array(self.start_position) 

    if (self.agent_position == self.negative_reward2).all():
      reward = -10  
    
    if (self.agent_position == self.reward1_position).all():
      if not self.reward1_done:
        reward = 20
        self.reward1_done = True
    
    if (self.agent_position == self.reward2_position).all():
      if not self.reward2_done:
        reward = 10
        self.reward2_done = True
    
    if (self.agent_position == self.goal_position).all():
      reward = 30

    self.timestep += 1
    if (self.agent_position == self.temp).all():
       reward = -10

    if (self.agent_position == self.goal_position).all() or self.timestep >= self.max_timesteps:
      done = True
    else:
      done = False
    info = {}
    return self.agent_position, reward, done, info

  def render(self):
    self.state = np.zeros((4, 4))
    self.state[tuple(self.negative_reward1)] = 0.3
    self.state[tuple(self.negative_reward2)] = 0.2
    self.state[tuple(self.reward1_position)] = 0.6
    self.state[tuple(self.reward2_position)] = 0.5
    self.state[tuple(self.goal_position)] = 0.7
    self.state[tuple(self.agent_position)] = 1
    plt.figure()
    plt.imshow(self.state)   
```
